# DATA

The data is  collected for group project of data science for sustainability class for years 2013 to 2021, which is semi cleaned by my group mates. This data is used and truncated the years 2019-2021 and columns which are required for this project. This data is saved in output folder consisting two datasets one outage data and other weather data for further merging in ArcGIS pro and python

```python
df = pd.read_csv('../Data/cleaned_premerge/outage_data_final.csv' )
df.head(5)
```

```python
df = df.loc[df['year'] > 2018]
```

```python
df.columns
```

```python
# taking only required columns
df_new = df[['date_out', 'time_out', 'day', 'month', 'year', 'date_in',
       'time_in', 'outage_duration', 'number_of_customers_affected',
       'city_town','reason_for_outage','Latitude', 'Longitude']].copy()
df_new.head(5)
```

```python
percent_missing = df_new.isnull().sum() * 100 / len(df_new)
missing_value_df = pd.DataFrame({'column_name': df_new.columns,
                                 'percent_missing': percent_missing})
missing_value_df
```

```python
#saving the data
df_new.to_csv('../output/outage_final.csv', index=False)
```

```python
#weather data
df = pd.read_csv('../Data/cleaned_premerge/weather_data_station.csv' )
df.head(5)
```

```python
df = df.loc[df['YEAR'] > 2018]
```

```python
df.columns
```

```python
# taking only required columns
df_new = df[['STATION','NAME', 'LATITUDE', 'LONGITUDE', 'DATE', 'AWND', 'PRCP', 'SNOW', 'TAVG',
       'TMAX', 'TMIN','WSF2','YEAR', 'MONTH', 'DAY']].copy()
df_new.head(5)
```

```python
df_new = df_new.rename(columns={'AWND':'Avg_daily_wind', 'PRCP':'precipitation', 'SNOW':'snow',
       'TMAX':'temp_max', 'TMIN':'temp_min','WSF2':'fastest_2_min_wind_speed'})
df_new.head(5)
```

```python
percent_missing = df_new.isnull().sum() * 100 / len(df_new)
missing_value_df = pd.DataFrame({'column_name': df_new.columns,
                                 'percent_missing': percent_missing})
missing_value_df
```

```python
#saving the data
df_new.to_csv('../output/weather_final.csv', index=False)
```
